# Background

The business model of sports

To begin, let's go through the fundamentals. In the most basic sense, sports leagues earn money from three sources: broadcasting (media rights sales), commercial (sponsorship and advertising agreements), and match day revenue (ticketing and hospitality).

Professional sports leagues are similar to entertainment corporations in that each team in the league is a separate channel. The leagues define the overall 'programming schedule' (the regulations of the game and the fixture list), but the teams have their own identities, personnel, and fan bases. This parallel may irritate purists, but the reality is that the more eyes on a product, the more valuable it is in both sports and entertainment.

As evidenced by revenue data from the main leagues over the last five years, all major sports are reliant on TV money. The global value of sports media rights is roughly $50 billion, yet only ten sports leagues account for 60% of it.

If the closure lasts longer than a few days, the leagues will be unable to meet their obligations to broadcasters, limiting their capacity to transfer revenue back to the teams. # Modelling

From our data we want to study the various patterns occured during the pandemic and study them to test our hypothesis "Covid Affected Sports in Terms of Revenue and Audience". So we implemented Segmentation techniques and observe our findings.Our initial Model is K-Means Clustering

K-Means Clustering

```python
from sklearn.cluster import KMeans 
```

We imported KMeans from sklearn.cluster.

```python
from sklearn.metrics import silhouette_score
n_clusters = [2,3,4,5,6,7,8,9,10] # number of clusters
clusters_inertia = [] # inertia of clusters
s_scores = [] # silhouette scores

for n in n_clusters:
    KM_est = KMeans(n_clusters=n, init='k-means++').fit(X_numerics)
    clusters_inertia.append(KM_est.inertia_)    # data for the elbow method
    silhouette_avg = silhouette_score(X_numerics, KM_est.labels_)
    s_scores.append(silhouette_avg) # data for the silhouette score method
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax = sns.lineplot(n_clusters, clusters_inertia, marker='o', ax=ax)
ax.set_title("Elbow method")
ax.set_xlabel("number of clusters")
ax.set_ylabel("clusters inertia")
ax.axvline(5, ls="--", c="red")
ax.axvline(6, ls="--", c="red")
plt.grid()
plt.show()
```

The above graph shows the Elbow method and ideal number of Clusters, in our case its 5 or 6

```python
fig, ax = plt.subplots(figsize=(12,5))
ax = sns.lineplot(n_clusters, s_scores, marker='o', ax=ax)
ax.set_title("Silhouette score method")
ax.set_xlabel("number of clusters")
ax.set_ylabel("Silhouette score")
ax.axvline(6, ls="--", c="red")
plt.grid()
plt.show()
```

From Silhouette Score method we found the ideal number of clusters = 6

The number of observations in our clusters range from 72 to 333.

Let's look at some statistics within our clusters

```python
KM_6_clusters = KMeans(n_clusters=6, init='k-means++').fit(X_numerics) # initialise and fit K-Means model

KM6_clustered = X_numerics.copy()
KM6_clustered.loc[:,'Cluster'] = KM_6_clusters.labels_ # append labels to points
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

We imported the required libraries of DBScan and metrics from sklearn.learn cluster and sklearn respectively.

```python
df_num = StandardScaler().fit_transform(data)
```

```python
A = []
B = []
C = []

for i in np.linspace(0.1,5,50):
    db = DBSCAN(eps=i, min_samples=10).fit(df_num)

    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    
    sum = 0
    for t in labels:
        if t == -1: 
            sum = sum + 1
    C.append(sum)
    
    
      
    A.append(i)
    B.append(int(n_clusters_))
```

labels are the label of the clusters. If the label is -1, then the observation is an outlier/noise within our dataset.

db.core_sampleindices are the indices of the core points in the cluster, the indices that are excluded here are of outliers and the edges of the clusters.

```python
results = pd.DataFrame([A,B,C]).T
results.columns = ['distance','Number of clusters','Number of outliers']
results.plot(x='distance',y='Number of clusters',figsize=(10,6))
```

Based on the above plot, I decided to go forward with a distance (epsilon) value of 2

```python
db = DBSCAN(eps=2, min_samples=10).fit(df_num)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_clusters_
```

The number of clusters in our dataset is 3

```python
data = data.join(pd.DataFrame(labels))
data = data.rename(columns={0:'Cluster'})

data['Cluster'].value_counts()
```

# Hypothesis Testing

```python
H0 = 'Covid 19 does not have an impact on sports in terms of revenue and audience'
H1 = 'Covid 19 has an impact on sports in terms of revenue and audience'
```

```python
alpha = 0.05
```

```python
from scipy.stats import shapiro
p = round(shapiro(data)[1], 2)
```

```python
if p > alpha:
  print(f"{p} > {alpha}. We fail to reject Null Hypothesis. {H0}")
else:
  print(f"{p} <= {alpha}. We reject Null Hypothesis. {H1}")
```
